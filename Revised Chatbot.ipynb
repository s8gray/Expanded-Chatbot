```python
#Import extra code to use 
import random
import string
import nltk
```

```python
#Test if input is a question
def is_question(input_string):
    
    output=bool
    
    if '?' in input_string:
        output=True   
    
    else:
        output=False
    
    return output
```

```python
#If the input has punctuation, remove it
def remove_punctuation(input_string):
    
    out_string=''
    
    for char in input_string:
        
        if char not in string.punctuation:
            out_string=out_string+char
    
    return out_string
```

```python
#Prepare the text all together
def prepare_text(input_string):
    
    out_list=[]
    
    for char in input_string: 
        
        temp_string = remove_punctuation(input_string) #remove punctuation, assign as temp
        
        temp_string = temp_string.lower() #take old temp, make lowercase
        
        out_list = temp_string.split() #split temp into list

    return out_list
```

```python
#repeat what is said 
def respond_echo(input_string, number_of_echoes, spacer):
    echo_output = ''
    
    if input_string is not None:
        echo_output = (input_string + spacer) * number_of_echoes
    else:
        echo_output=None
    return echo_output
```

```python
#select an outpuut
def selector(input_list, check_list, return_list):
    output=None
    for item in input_list:
        if item in check_list:
            output=random.choice(return_list)
            break
    return output
```

```python
#Put strings together
def string_concatenator(string1, string2, seperator):
    output=string1+seperator+string2
        
    return output
```

```python
#Change from list to string
def list_to_string(input_list, separator):
    print(input_list)
    output=input_list[0]
    for item in input_list[1:]:
        output=string_concatenator(output, item, separator)
    return output
```

```python
#Id user wants to end chat
def end_chat(input_list):
    output = bool
    if 'quit' in input_list:
        output = True
    else:
        output = False
    return output
```

```python
def is_in_list(list_one, list_two):
    """Check if any element of list_one is in list_two."""
    
    for element in list_one:
        if element in list_two:
            return True
    return False

def find_in_list(list_one, list_two):
    """Find and return an element from list_one that is in list_two, or None otherwise."""
    
    for element in list_one:
        if element in list_two:
            return element
    return None
```

```python
#Function to allow user to pick a chatbot
def choose():
    """Main function to run our chatbot."""
    
    print('Choose a bot: optimist or pessimist')
    print('Make the right choice or your computer will crash. And Im not paying for that.')
    
    start = input('INPUT :\t') 
    
    if 'pessimist' in start.lower():
        print('I am the pessimist bot, and life sucks. Why am I here?')
        sad_chat()
        return
   
    elif 'optimist' in start.lower():
        print('I am the optimist bot, and Im so happy to be able to talk to you today!!.')
        happy_chat()
        return
    
    else:
        print('I am the regular bot.')
        regular_chat()
        return
```

```python
#Main function for regular chatbot
#This code comes from A3
def regular_chat():
    chat = True
    while chat:

        # Get a message from the user
        reg_msg = input('INPUT :\t')
        out_msg = None

        # Check if the input is a question
        question = is_question(reg_msg)

        # Prepare the input message
        reg_msg = prepare_text(reg_msg)
        
        # Check for an end msg 
        if end_chat(reg_msg):
            out_msg = 'Bye!'
            chat = False

        # Check for a selection of topics that we have defined to respond to
        #   Here, we will check for a series of topics that we have designed to answer to
        if not out_msg:

            # Initialize to collect a list of possible outputs
            outs = []

            # Check if the input looks like a greeting, add a greeting output if so
            outs.append(selector(reg_msg, GREETINGS_IN, GREETINGS_OUT))

            # Check if the input looks like a computer thing, add a computer output if so
            outs.append(selector(reg_msg, COMP_IN, COMP_OUT))

            # Check if the input mentions a person that is specified, add a person output if so
            if is_in_list(reg_msg, PEOPLE_IN):
                name = find_in_list(reg_msg, PEOPLE_IN)
                outs.append(list_to_string([PEOPLE_NAMES[name], name.capitalize(),
                                            selector(reg_msg, PEOPLE_IN, PEOPLE_OUT)], ' '))

            # Check if the input looks like a joke, add a repeat joke output if so
            outs.append(respond_echo(selector(reg_msg, JOKES_IN, JOKES_OUT), 3, ''))

            # Check if the input has some words we don't want to talk about, say that, if so
            if is_in_list(reg_msg, NONO_IN):
                outs.append(list_to_string([selector(reg_msg, NONO_IN, NONO_OUT), find_in_list(reg_msg, NONO_IN)], ' '))

            # IF YOU WANTED TO ADD MORE TOPICS TO RESPOND TO, YOU COULD ADD THEM IN HERE

            # We could have selected multiple outputs from the topic search above (if multiple return possible outputs)
            # We also might have appended None in some cases, meaning we don't have a reply
            # To deal with this, we are going to randomly select an output from the set of outputs that are not None
            options = list(filter(None, outs))
            if options:
                out_msg = random.choice(options)

        # If we don't have an output yet, but the input was a question, return msg related to it being a question
        if not out_msg and question:
            out_msg = QUESTION

        # Catch-all to say something if msg not caught & processed so far
        if not out_msg:
            out_msg = random.choice(UNKNOWN)

        print('OUTPUT:', out_msg)
```
